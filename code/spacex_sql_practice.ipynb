### Task 1




##### Display the names of the unique launch sites  in the space mission


```python
cur.execute('''
    SELECT DISTINCT Launch_Site
    FROM SPACEXTBL;
''')

query = cur.fetchall()

print("Names of the unique launch sites in the space mission:")
for row in query:
    print( row[0])
```


### Task 2


#####  Display 5 records where launch sites begin with the string 'CCA' 


```python
cur.execute('''
    SELECT * 
    FROM SPACEXTBL
    WHERE Launch_Site LIKE 'CCA%'
    LIMIT 5;
''')

query = cur.fetchall()

print("Display 5 records where launch sites begin with the string 'CCA':")
for row in query:
    print(row,'\n')
```

### Task 3




##### Display the total payload mass carried by boosters launched by NASA (CRS)


```python
cur.execute('''
    SELECT SUM(PAYLOAD_MASS__KG_) AS total_payload_mass
    FROM SPACEXTBL
    WHERE Customer = 'NASA (CRS)';
''')

query = cur.fetchall()

print("Display the total payload mass carried by boosters launched by NASA (CRS):")
for row in query:
    print(row[0])
```

### Task 4




##### Display average payload mass carried by booster version F9 v1.1


```python
cur.execute('''
    SELECT AVG(PAYLOAD_MASS__KG_) AS average_payload_mass
    FROM SPACEXTBL
    WHERE Booster_Version = 'F9 v1.1';
''')

query = cur.fetchall()

print("Display average payload mass carried by booster version F9 v1.1:")
for row in query:
    print(row[0])
```

### Task 5

##### List the date when the first succesful landing outcome in ground pad was acheived.


_Hint:Use min function_ 


```python
cur.execute('''
    SELECT MIN(Date)
    FROM SPACEXTBL
    WHERE Landing_Outcome = 'Success (ground pad)';
''')

query = cur.fetchall()

print("List the date when the first succesful landing outcome in ground pad was acheived.:")
for row in query:
    print(row[0])
```

### Task 6

##### List the names of the boosters which have success in drone ship and have payload mass greater than 4000 but less than 6000


```python
cur.execute('''
    SELECT Booster_Version
    FROM SPACEXTBL
    WHERE Landing_Outcome = 'Success (drone ship)'
    AND PAYLOAD_MASS__KG_ BETWEEN 4000 AND 6000;
''')

query = cur.fetchall()

print("List the names of the boosters which have success in drone ship and have payload mass greater than 4000 but less than 6000:")
for row in query:
    print(row[0])
```

### Task 7




##### List the total number of successful and failure mission outcomes


```python
cur.execute('''
    SELECT 
        SUM(CASE WHEN Landing_Outcome LIKE '%Success%' THEN 1 ELSE 0 END) AS count_success,
        SUM(CASE WHEN Landing_Outcome LIKE '%Failure%' THEN 1 ELSE 0 END) AS count_failure
    FROM SPACEXTBL;
''')

query = cur.fetchall()

print("List the total number of successful and failure mission outcomes:")
for row in query:
    print("Success:", row[0], " | Failure:", row[1])
```

### Task 8



##### List all the booster_versions that have carried the maximum payload mass, using a subquery with a suitable aggregate function.


```python
cur.execute('''
    SELECT Booster_Version 
    FROM (
        SELECT Booster_Version, MAX(PAYLOAD_MASS__KG_) AS max_payload_mass
        FROM SPACEXTBL
        GROUP BY Booster_Version
    )
    WHERE max_payload_mass = (
        SELECT MAX(PAYLOAD_MASS__KG_)
        FROM SPACEXTBL
    );
''')

query = cur.fetchall()

print("List all the booster_versions that have carried the maximum payload mass, using a subquery with a suitable aggregate function.:")
for row in query:
    print(row[0])
```

### Task 9


##### List the records which will display the month names, failure landing_outcomes in drone ship ,booster versions, launch_site for the months in year 2015.

**Note: SQLLite does not support monthnames. So you need to use  substr(Date, 6,2) as month to get the months and substr(Date,0,5)='2015' for year.**


```python
cur.execute('''
    SELECT CASE substr(Date, 6, 2)
            WHEN '01' THEN 'January'
            WHEN '02' THEN 'February'
            WHEN '03' THEN 'March'
            WHEN '04' THEN 'April'
            WHEN '05' THEN 'May'
            WHEN '06' THEN 'June'
            WHEN '07' THEN 'July'
            WHEN '08' THEN 'August'
            WHEN '09' THEN 'September'
            WHEN '10' THEN 'October'
            WHEN '11' THEN 'November'
            WHEN '12' THEN 'December'
        END AS month_name,
        Landing_Outcome, 
        Booster_Version, 
        Launch_Site
    FROM SPACEXTBL
    WHERE Landing_Outcome = 'Failure (drone ship)' 
        AND substr(Date, 0, 5)='2015';
''')

query = cur.fetchall()

print("List the records which will display the month names, failure landing_outcomes in drone ship ,booster versions, launch_site for the months in year 2015.:")
for row in query:
    print(row)
```

### Task 10




##### Rank the count of landing outcomes (such as Failure (drone ship) or Success (ground pad)) between the date 2010-06-04 and 2017-03-20, in descending order.


```python
cur.execute('''
    SELECT Landing_Outcome, count_landing_outcome
    FROM(
        SELECT Landing_Outcome, COUNT(Landing_Outcome) AS count_landing_outcome
        FROM SPACEXTBL
        WHERE Date BETWEEN '2010-06-04' AND '2017-03-20'
        GROUP BY Landing_Outcome 
    )
    WHERE Landing_Outcome = 'Failure (drone ship)' 
       OR Landing_Outcome = 'Success (ground pad)'
    ORDER BY count_landing_outcome DESC
''')

query = cur.fetchall()

print("Rank the count of landing outcomes (such as Failure (drone ship) or Success (ground pad)) between the date 2010-06-04 and 2017-03-20, in descending order.:")
for row in query:
    print(row)
```
